## 프로그래머스 - K번째수
#### 문제 설명
배열 array의 i번째 숫자부터 j번째 숫자까지 자르고 정렬했을 때, k번째에 있는 수를 구하려 합니다.

예를 들어 array가 [1, 5, 2, 6, 3, 7, 4], i = 2, j = 5, k = 3이라면

array의 2번째부터 5번째까지 자르면 [5, 2, 6, 3]입니다.

1에서 나온 배열을 정렬하면 [2, 3, 5, 6]입니다.

2에서 나온 배열의 3번째 숫자는 5입니다.

배열 array, [i, j, k]를 원소로 가진 2차원 배열 commands가 매개변수로 주어질 때, commands의 모든 원소에 대해 앞서 설명한 연산을 적용했을 때 나온 결과를 배열에 담아 return 하도록 solution 함수를 작성해주세요.



---


#### 제한사항
array의 길이는 1 이상 100 이하입니다.

array의 각 원소는 1 이상 100 이하입니다.

commands의 길이는 1 이상 50 이하입니다.

commands의 각 원소는 길이가 3입니다.



---
|array|commands|return|
|------|---|---|
|[1, 5, 2, 6, 3, 7, 4]|[[2, 5, 3], [4, 4, 1], [1, 7, 3]]|[5, 6, 3]|

```
# bubble_sort를 이용한 풀이
# bubble_sort: 이웃한 두 값을 비교하여 정렬
# early stopping? 어디에서 처리해주냐? -> flag 안 넣어줘도되나?
def bubble_sort(array) :  
    for size in reversed(range(len(array))) :
        flag = 0
        for i in range(size) : 
            if array[i] > array[i+1] : 
                array[i], array[i+1] =  array[i+1], array[i]
                flag = 1

        if flag == 0 : return array

    return array 

def solution(array, commands):
    answer = []
    
    for c in commands :
        sorted = bubble_sort(array[c[0]-1:c[1]])
        print(sorted, c[2])
        answer.append(sorted[c[2]-1])
        
    return answer

a = solution([1, 5, 2, 6, 3, 7, 4], [[2, 5, 3], [4, 4, 1], [1, 7, 3]]	)
print(a)
```

## 프로그래머스 - 가장 큰 수

#### 문제 설명
0 또는 양의 정수가 주어졌을 때, 정수를 이어 붙여 만들 수 있는 가장 큰 수를 알아내 주세요.

예를 들어, 주어진 정수가 [6, 10, 2]라면 [6102, 6210, 1062, 1026, 2610, 2106]를 만들 수 있고, 이중 가장 큰 수는 6210입니다.

0 또는 양의 정수가 담긴 배열 numbers가 매개변수로 주어질 때, 순서를 재배치하여 만들 수 있는 가장 큰 수를 문자열로 바꾸어 return 하도록 solution 함수를 작성해주세요.

#### 제한 사항
numbers의 길이는 1 이상 100,000 이하입니다.
numbers의 원소는 0 이상 1,000 이하입니다.
정답이 너무 클 수 있으니 문자열로 바꾸어 return 합니다.

#### 입출력 예
|numbers|return|
|------|---|
|[6, 10, 2]|6210|
|[3, 30, 34, 5, 9]|9534330|

```
# 다른 풀이
# 최대 max len 구해서 각 숫자들을 최대 max로 맞춰주기
# ex) 324, 3,39 -> 324, 333, 393 그 다음에 sort 
# 왜 안되는거야!!

# 12, 121일 때 안된다 1212 1211 이렇게 반복해줘야한다. 4자리수로 만들자
# 1211 1211 -> 구별이 되지 않음. 1212 1211로 해줘야 끝자리까지 구별이 가능하다
# 하지만 아직도 테스트 10이 안됨

# str 정렬 사용
# s = [102, 111, 222, 272, 10000]의 케이스에서
# str 정렬 [10000, 102, 111, 222, 272]
# int 정렬 [102, 111, 222, 272, 10000]
# 아마 5자리를 벗어나는 경우에서 sort가 잘 안됬던거 같다.
# -> 또 블로그 풀이 보니깐 노상관인거같은데?

# 10번째 케이스말고 다 된다, 10번째는 왜 안되는겨

def change_num_as_maxlen(numbers) :
    result = [] # [[number, idx of adding]]
    for i in range(len(numbers)) : 
        result.append([numbers[i], numbers[i]*3])

    return result

def remove_added_num(numbers_n_idx) : 
    result = []

    for i in range(len(numbers_n_idx)) :
        result.append(numbers_n_idx[i][0])

    return result

## 두자리수 여러개면 안됨
def solution(numbers):
    numbers = list(map(str, numbers))

    numbers = change_num_as_maxlen(numbers)
    numbers   = sorted(numbers, reverse=True, key=lambda x : x[1])
    numbers = remove_added_num(numbers)

    result = "".join(numbers)

    return result


# 알게 된 점
# sorted 함수 사용법
# print(sorted(values, key=str))  -> 반환타입 values의 원래 type
# print(sorted(map(str, values))) -> 반환타입 str

# max 길이 찾는법
# max(list, key = str)

# 조건부 sorting
# sorted(a, key = lambda x : x[0])
# 출처: https://dailyheumsi.tistory.com/67 [하나씩 점을 찍어 나가며]
```

## 프로그래머스 H-index

### 문제 설명
H-Index는 과학자의 생산성과 영향력을 나타내는 지표입니다. 어느 과학자의 H-Index를 나타내는 값인 h를 구하려고 합니다. 위키백과1에 따르면, H-Index는 다음과 같이 구합니다.

어떤 과학자가 발표한 논문 n편 중, h번 이상 인용된 논문이 h편 이상이고 나머지 논문이 h번 이하 인용되었다면 h의 최댓값이 이 과학자의 H-Index입니다.

어떤 과학자가 발표한 논문의 인용 횟수를 담은 배열 citations가 매개변수로 주어질 때, 이 과학자의 H-Index를 return 하도록 solution 함수를 작성해주세요.

### 제한사항
과학자가 발표한 논문의 수는 1편 이상 1,000편 이하입니다.
논문별 인용 횟수는 0회 이상 10,000회 이하입니다.

###입출력 예

|citations|return|
|------|---|
|[3, 0, 6, 1, 5]|3|

```
# h번 이상 인용된 논문이 h편 이상 나머지 논문이 h번 이하 인용
# 역순 정렬 후에 h번을 세고 비교하면서
# 6 5 3 1 0 앞에서 부터 0, len-idx > h, idx < h
# 8 8 7 5 4 2 0

# 문제가 이 방법이 아니다.
# 논문이 인용된 횟수 >= 인용된 논문의 개수

def solution(citations):
    citations.sort(reverse=True)
    n = len(citations)

    for i, c in enumerate(citations) :
        if i >= c : 
            return i
    
    return n

#solution([3, 0, 6, 1, 5])
solution([0]) # 이런 경우에 안되는데
```

```
# https://m.blog.naver.com/PostView.nhn?blogId=promarketyj&logNo=221434899288&categoryNo=22&proxyReferer=https:%2F%2Fwww.google.com%2F
# h번 이상 인용된 논문이 h개 이상인 경우
# 예시)
# 10 7 4 3 2 1 의 논문 인용 수 인 경우
# 0  1 2 3 4 5
# h는 3이다. 3번 이상 인용된 논문이 3개이기 때문에

# 5 5 5 5 5
# 0 1 2 3 4 
# h는 5이다. 5번 이상 인용된 논문이 5개이기 때문에
# 내림차순으로 정렬 후, idx가 논문 인용된 회수보다 커지면 idx를 return하면 된다.

# 단, [0, 0, 0, 0]의 경우 [5, 5, 5, 5]의 경우가 수행되지 않음
def solution(citations):
    citations.sort(reverse=True)

    for idx, citation in enumerate(citations) :
        if idx >= citation :
            return idx
    return idx+1

#solution([5, 5, 5, 5, 5])
#solution([0, 0, 0, 0]) # 이런 경우에 안되는데
solution([0,1,5,6,9])
```
