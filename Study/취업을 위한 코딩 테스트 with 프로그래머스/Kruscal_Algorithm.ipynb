```
# 홍빈 언니 풀이
def getParent(parent,n):
    while (parent[n]!= n):#집합의 루트가 아니라면 반복.
        parent[n]=parent[parent[n]]# 집합의 루트를 찾기 위해 2개씩 건너 뛰면서 부모 갱신
        n=parent[n]# 2개 건너 뛴 부모를 자기 자신으로 갱신
    return n #루트를 찾았다면 루트 반환

def union(parent,a,b):
    a=getParent(parent,a)
    b=getParent(parent,b)
    
    if(a<b):
        parent[b]=a
    else:
        parent[a]=b

def solution(n, costs):
    answer=0
    bridge=0
    parent=[i for i in range(0,n+1)]
    
    costs.sort(key=lambda x:x[2]) # ascending sort 
    
    for cost in costs:
        if(getParent(parent,cost[0])!=getParent(parent,cost[1])):
            union(parent,cost[0],cost[1])
            bridge +=1
            answer+=cost[2]
            
        if(bridge==n-1):
            break
    
    return answer
```
